# Steps to build a Model in Python
## 1) Import your libraries
## 2) Imprt your data
## 3) Train a model (today we train a Tree)
## 4) Test your model (Visualize it)

# 1) Import required libraries

```python
# Machine Learning Packages
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO 
#Visualization Packages
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

# Iris Flowers

```python
Image(filename='/Users/jay.ghorbani/Desktop/iris.jpeg')
```

# 2) Import the data

```python
iris=datasets.load_iris()
print(iris.target_names)
flower_data=pd.DataFrame(iris.data, columns=iris.feature_names)
flower_names=iris.target 
# ['setosa' -->0 'versicolor'-->1 'virginica'-->2]
```

# 3) Train a tree classifier 

```python
tree=DecisionTreeClassifier()
tree.fit(flower_data,flower_names)
```

# 4) Visualize the tree

```python
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data,filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
# ['setosa' --> Orange 
#  'versicolor'-->Green 
#  'virginica'-->Purple ]
```

# Predict a flower type

```python
#           [X0 ,  X1, X2,   X3 ]
test_flower=[0.1, 0.5,  1,  1]

tree.predict([test_flower])
# ['setosa' -->0 'versicolor'-->1 'virginica'-->2]
```
